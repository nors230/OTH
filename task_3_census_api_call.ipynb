```python
#for merging purposes
import requests
request_obj = requests.get('https://api.census.gov/data/2017/acs/acs5/?get=B01003_001E,B07011_001E,B15003_001E,B23025_005E,B17001_001E&for=state:*')
json = request_obj.json()
df_zip_info = pd.DataFrame(json)
df_zip_info.columns = ['total_population', 'median_income', 'education_attainment', 'employment_status', 'poverty_status', 'zipcode']

#df_zip_info['per_bl'] = df_zip_info['total_black'] / df_zip_info['total_population']

df_zip_info

#still need to id the most useful metrics and apply properly
```
